```python
%load_ext autoreload
%autoreload 2
    
from code.multiarmed_bandits import *
from tqdm import tqdm
```

```python
NSTEPS = 1000
NREPEAT = 1000
NBANDITS = 2
```

```python
bandit_means = [np.random.randn() for n in range(NBANDITS)]
```

```python
print(bandit_means)
```

```python
datas = []
for i in tqdm(range(NREPEAT)):
    bandits, data = run_experiment(bandit_means, 1.0, NSTEPS, rule="rate")
    datas.append(data.sum())

_ = plt.hist(np.array(datas), bins=10, alpha=0.5, label=f"eps: {np.array(datas).mean():.2f}")

datas = []
for i in tqdm(range(NREPEAT)):
    bandits, data = run_experiment(bandit_means, 0.5, NSTEPS, rule="epsilon")
    datas.append(data.sum())

_ = plt.hist(np.array(datas), bins=10, alpha=0.5, 
             label=f"eps_geom: {np.array(datas).mean():.2f}")

datas = []
for i in tqdm(range(NREPEAT)):
    bandits, data = run_experiment(bandit_means, 0.5, NSTEPS, rule="none")
    datas.append(data.sum())
_ = plt.hist(np.array(datas), bins=10, alpha=0.1,
             label=f"0.1: {np.array(datas).mean():.2f}")
datas = []
for i in tqdm(range(NREPEAT)):
    bandits, data = run_experiment(bandit_means, 0.5, NSTEPS, rule="none")
    datas.append(data.sum())
_ = plt.hist(np.array(datas), bins=10, alpha=1,
             label=f"1: {np.array(datas).mean():.2f}")

plt.legend()
```
